```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
th_list = []
tr_list = []
td_list = []

url = 'https://www.numbeo.com/quality-of-life/rankings_by_country.jsp'
r = requests.get(url)

soup = BeautifulSoup(r.text, 'html.parser')

thead = soup.find('thead')

ths = thead.find_all('th')

for th in ths:
    th_list.append(th.text)

tbody = soup.find('tbody')

trs = tbody.find_all('tr')

for tr in trs:
    td_list = []
    tds = tr.find_all('td')
    for td in tds:
        td_list.append(td.text)
    tr_list.append(td_list)
```

```python
df = pd.DataFrame(tr_list, columns=th_list)
df
```

```python
countries = []
distances = []

url = 'https://www.distancefromto.net/distance-from-poland-country'
r = requests.get(url)

soup = BeautifulSoup(r.text, 'html.parser')

table = soup.find('table')

tds = table.find_all('td')

for i, td in enumerate(tds):
  if i % 3 == 0:
    countries.append(td.text.replace('Distance from ', '').replace(' to Poland', '').replace('Poland to ', ''))
  if i % 3 == 1:
    distances.append(td.text.replace(' km', ''))
```

```python
df2 = pd.DataFrame()
df2['Country'] = countries
df2['Distance'] = distances
df2
```

```python
df_merged = df.merge(df2,on='Country',how="left")
df_merged['Rank'] = range(1,88)
df_merged = df_merged.set_index('Rank')
df_merged
```

```python
df_merged['Distance'] = df_merged['Distance'].astype(str).apply(lambda x: x.replace(',',''))
df_merged
```

```python
cols = df_merged.columns[1:]
df_merged[cols] = df_merged[cols].apply(pd.to_numeric, errors='coerce')
df_merged
```

```python
df_merged = df_merged[df_merged['Quality of Life Index'] > 132]
df_merged = df_merged[df_merged['Purchasing Power Index'] > 60]
df_merged = df_merged[df_merged['Safety Index'] > 60]
df_merged = df_merged[df_merged['Health Care Index'] > 50]
df_merged = df_merged[df_merged['Property Price to Income Ratio'] < 14]
df_merged = df_merged[df_merged['Climate Index'] > 50]
df_merged
```

```python
df_merged.to_excel('life_quality_index.xlsx')
```
